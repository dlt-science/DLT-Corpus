# Publications dataset

```python
import sys
sys.path.append('./../')
from src.models.continual_training import load_dataset_safe

import multiprocessing

# Determine number of processes to use based on CPU count
num_proc = max(1, int(1 / 3 * multiprocessing.cpu_count()))

# Cache directory for datasets
hf_datasets = "./../data/hf_datasets"

# Manually reviewed datasets
manually_reviewed_dir = "./../data/manual_revision"
```

```python
# Cleans the publication dataset
publications_ds = load_dataset_safe(hf_datasets, "ExponentialScience/DLT-Literature")
```

Filter the dataset for a manual revision of the data:

```python
import os



publications_df = publications_ds.select_columns(["year", "title", "url", "paperId"]).to_pandas()
publications_df[publications_df["year"] <= 2008].sort_values("year").to_csv(os.path.join(manually_reviewed_dir, "earliest_papers.csv"), index=False)
publications_df[publications_df["year"] > 2008].sort_values("year").to_csv(os.path.join(manually_reviewed_dir, "latest_papers.csv"), index=False)
```

Load manually revised data:

```python
import pandas as pd
import os

df_1 = pd.read_csv(os.path.join(manually_reviewed_dir, "latest_papers_filtered.csv"))
df_2 = pd.read_csv(os.path.join(manually_reviewed_dir, "earliest_papers_filtered.csv"))

# Combine the two DataFrames
combined_df = pd.concat([df_1, df_2], ignore_index=True)

# Filter the list of only the papers to remove based on manual revision
papers_to_remove = combined_df[combined_df['remove'] == 1]['paperId'].tolist()

papers_to_remove
```

Filter out the list of paperIDs that were manually revised from the original dataset:

```python
print(f"Number of papers to remove: {len(papers_to_remove)}")
print(f"Total number of papers before cleaning: {len(publications_ds)}")
```

```python
publications_ds = publications_ds.filter(lambda example: example['paperId'] not in papers_to_remove, 
                                         load_from_cache_file=True, 
                                         num_proc=num_proc)
```

```python
print(f"Number of papers after cleaning: {len(publications_ds)}")
```

```python
# Push dataset to hub
publications_ds.push_to_hub("ExponentialScience/DLT-Literature", private=True)
```
